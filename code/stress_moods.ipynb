```python
import pandas as pd
import numpy as np
import pickle
from datetime import datetime
```

```python
data = pickle.load(open('../data/moods_data.p','rb'))
```

```python
data['window'].loc[0]['start']
```

```python
def get_day(a):
    return a['start'].strftime('%Y-%m-%d')

data['day'] = data['window'].apply(lambda a:get_day(a))
```

```python
import matplotlib.pyplot as plt
%matplotlib notebook
def plot_day(data):
    data = data.sort_values('timestamp1').reset_index(drop=True)
    plt.figure()
    plt.plot(data['timestamp1'],data['probability'],'*-r')
    data = data.sort_values('timestamp').reset_index(drop=True)
    plt.plot(data['timestamp'],data['stress_likelihood'],'o-b')
    plt.show()
    
data_all = [a for i,a in data.groupby(['user','day'])]
```

```python
a = [plot_day(a) for a in data_all if a.dropna().shape[0]>200]
```
